### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Import libaries

```python
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
# Train Test Split
from sklearn.model_selection import train_test_split
# To fit the model (training and prediction)
from sklearn.cluster import KMeans

# Fro classification report (model evaluation)
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import roc_auc_score
```

### Prepare the Dataset

```python
X, y = make_blobs(n_samples=400, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:,0], X[:,1])
```

  ### Use ELBOW METHOD TO find the k value

```python
K = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=350, n_init=10, random_state=0)
    kmeans.fit(X)
    K .append(kmeans.inertia_)
plt.plot(range(1, 11), K )
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('K ')
plt.show()
```

#### CAtegorize the data

```python
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
```

```python
#### PRediction
```
